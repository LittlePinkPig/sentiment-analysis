```python
from topic_model_util import tokenize, get_lemma, get_lemma2, prepare_text_for_lda
```

### Load data

```python
import pandas as pd

df = pd.read_csv('example_data/train.csv', index_col=0)
text_data = []
for row in df.itertuples():
    description = str(row.description).decode('utf-8')
    tokens = prepare_text_for_lda(description)
    text_data.append(tokens)
```

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
```

```python
import os
import pickle

if not os.path.exists('out'):
    os.mkdir('out')

pickle.dump(corpus, open('out/corpus.pkl', 'wb'))
dictionary.save('out/dictionary.gensim')
```

### Train models for 3, 5, 10 topics

```python
import gensim

    
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=3, id2word=dictionary, passes=15)
ldamodel.save('out/model3.gensim')
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=5, id2word=dictionary, passes=15)
ldamodel.save('out/model5.gensim')
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)
ldamodel.save('out/model10.gensim')
```

### visualization

```python
import pyLDAvis.gensim

dictionary = gensim.corpora.Dictionary.load('out/dictionary.gensim')
corpus = pickle.load(open('out/corpus.pkl', 'rb'))
```

```python
model3 = gensim.models.ldamodel.LdaModel.load('out/model3.gensim')
model3_display = pyLDAvis.gensim.prepare(model3, corpus, dictionary, sort_topics=True)
pyLDAvis.display(model3_display)
```
